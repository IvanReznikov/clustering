# Outlier detection for multiple clusters treated as multiple cluster

```python
import python_clustering
```

### Load Data

```python
mode = "per_class"

dataset = python_clustering.Dataset()
load_df = dataset.load("D31")
X = load_df.values
```

### Load Tasks Class

```python
tasks = python_clustering.Tasks()
```

### Run outlier detection

```python
result, detected_outliers, classifiers = tasks.detect_anomalies(X, methods = "all_besides_nn", 
                                outliers_fraction=0.05, mode=mode)
```

### Plot all methods classifications

```python
tasks.plot_individual_anomaly_classifiers(result, classifiers, detected_outliers, mode=mode)
```

### Plot overall outlier detection results

```python
tasks.plot_overall_anomaly_classifiers(result, classifiers, show_heatmap=True)
```
